### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'Number of students that scored 100 in Maths: {math_full}')
print(f'Number of students that scored 100 in Writing: {writing_full}')
print(f'Number of students that scored 100 in Reading: {reading_full}')
```

#####  Insights
 - From the above values we observe that more students excel in Reading. 
 - Only 7 students reached excelence in Math comparing to 17 in Reading.
 

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with score of 20 or lower in Maths: {math_less_20}')
print(f'Number of students with score of 20 or lower in Writing: {writing_less_20}')
print(f'Number of students with score of 20 or lower in Reading: {reading_less_20}')
```

#####  Insights
Based on the provided data, it is evident that students performed the poorest in Mathematics, while the Reading section showcased the highest level of achievement.
